```python
import matplotlib.pyplot as plt
```

```python
sales = [500,1000,700,800]
day = ['Mon','Tue','Wed','Thr']
```

```python
#plt.plot(x,y)
plt.plot(day,sales)
plt.xlabel("Days of the week")
plt.ylabel("Sales")
plt.title("Sales per day")
plt.grid()
plt.show()
```

Bar Plot

```python
# when you want to compare numerical and categorical values
# bivariate -- chart created for two variables
```

```python
products = ['A','B','C','D','E'] # categorical variable
sales = [200, 300, 250, 200, 400] # numerical variable

plt.figure(figsize = (8,5))
plt.bar(products, sales, color='red')
plt.xlabel("differnt products")
plt.ylabel("units sold")
plt.title("products vs sales")
plt.grid()
plt.show()
```

```python
fruits= ['Apple', 'Banana', 'Strawberry', 'Pappaya','Orange']
likes= [10, 20, 30, 20, 30]
```

```python
plt.bar(fruits, likes, width = 0.4, color = ['red','yellow','green','pink','orange'])
plt.xlabel("Fruits")
plt.ylabel("Percentage og people liking fruits")
plt.title("Fruits vs likes")
plt.grid()
plt.show()
```

```python
cars = ['Audi','BMW','Honda']
prices = [320000, 4000000, 500000]

plt.bar(cars, prices, color =['gray','black','blue'])
plt.show()
```

```python
# customer churn analysis

monthly_spend = [ 200, 150, 450,  700, 200, 500]

tenure = [1,2,3,4,4,5]
plt.figure(figsize = (5,5))
plt.scatter(tenure, monthly_spend)
plt.axvline(x=2.5, color='r', linestyle='--')
plt.axhline(y=300, color='r', linestyle='--')
plt.title("tenure vs monthly spend")
plt.xlabel("tenure")
plt.ylabel("monthly spend")
plt.grid()
plt.show()
```

## Heatmap

### Correlation

```python
# what is the relation of a column with another column
# tells you about the relationship between columns
# no_of_rooms, locality, area, parking--input cols and price - output
# exp, qualification, skills -- input cols and salary or job -- output
# how much one column is related to the other

# it gives a score between -1 to 1
```

```python
# +ve correlation

# if col1 increases and col2 also increases then there is +ve correlation
# if col2 decreases and col1 also decreases then there is +ve correlation

# study and marks
# temp and ice_cream sales
# customer vs sales
# sales and profit
# offers and purchases


# 0 - 1

# as much the score will be closer to +1 that much high +ve correlation will be there
```

```python
# -ve correlation

# if col1 inc col2 value dec
# if col1 dec col2 value inc


# exer, calories
# speed, time
# discounts, price
# demand and supply

# if your value is closer to -1 it means high -ve correlation is there
```

```python
# no correlation

# 0 or very close to 0

# sutdy hours and shoe size

# salary and area they are living
```
